```python
# MAIN
# Must be executed after started functions
import os
try:
    os.mkdir('.\\output') # Creates an output folder to analysed data
    print('Folder output created for finished files')
except:
    print('Folder output already exists')
    
# Open files to be analysed
filer = open('origem\\GastosDiretos.csv') # Read
filew = open('output\\gastos.txt','w') # Write
fileb = open('origem\\GastosDiretos.csv') # Read to counter


# Execute main function whitch convert CSV files into SQL
gastosPublicos()

# closes files opened fot Analysis
filer.close()
filew.close()
fileb.close()
```

```python
# MAIN FUNCTIONS

# Starts function that allows perform counting of elements to be analysed before analysis.
def contador(file):
    abc = 0
    asd = 0
    for row2 in file:
        asd = row2.split('\t')
        for a in asd:
            abc+=1
    return abc

# Main function to convert CSV files into SQL
def gastosPublicos():
    # Starting variables:
    counted = contador(fileb) # stores total itens to be analysed
    dataPgto = 0 # Stores the last paying date to be inserted in the function
    counter = 0 # Contador que permite checar em qual parte do procedimento o programa est√°
    idOrgaoSuperior = '' # Armazena o √∫ltimo ID do √ìrg√£o Superior para poder inserir na tabela como chave estrangeira
    idOrgaoSubordinado = '' # Armazena o √∫ltimo ID do √ìrg√£o Subordinado para poder inserir na tabela como chave estrangeira
    idFuncao = '' # Armazena o √∫ltimo ID da Fun√ß√£o para poder inserir na tabela como chave estrangeira
    idValorPago = 0 # Armazena o valor pago e posteriormente acrecenta a cada rodada para cria√ß√£o de ID
    idPrograma = '' # Armazena o √∫ltimo ID do Programa para poder inserir na tabela como chave estrangeira
    idSubFuncao = '' # Armazena o √∫ltimo ID da Subfun√ß√£o para poder inserir na tabela como chave estrangeira
    t1 = [] # Armazena todos os dados do √ìrg√£o Superior que j√° foram analizados para evitar inser√ß√µes duplas
    t2 = [] # Armazena todos os dados do √ìrg√£o Subordinado que j√° foram analizados para evitar inser√ß√µes duplas
    t3 = [] # Armazena todos os dados da Fun√ß√£o que j√° foram analizados para evitar inser√ß√µes duplas
    t4 = [] # Armazena todos os dados da Subfun√ß√£o que j√° foram analizados para evitar inser√ß√µes duplas
    t5 = [] # Armazena todos os dados do Programa que j√° foram analizados para evitar inser√ß√µes duplas
    
    for row in filer: # in√≠cio da aplica√ß√£o: para cada linha dentro do arquivo
        rowArr = row.split('\t') # Cria um vetor separando por cada tabula√ß√£o na linha
        i = 1 # Define o contador do pr√≥ximo loop
        for value in rowArr: # Para cada valor dentro do Vetor da Linha
            counter += 1 # Contador come√ßa a rodar
            if i == 1 or i == 2: # Se contador for 1 ou 2
                if value not in t1: # Se valor n√£o tiver sido gravado antes
                    t1.append(value) # Registra a grava√ß√£o do valor
                    if i == 1: # Se for 1 grava o valor com sintaxe MySQL no arquivo
                        idOrgaoSuperior = value
                        filew.write("INSERT INTO orgaosuperior (idOrgaoSuperior, nomeOrgaoSuperior) VALUES (")
                        filew.write(value)
                        filew.write(", '")

                    elif i == 2: # Se for 2 grava o valor com sintaxe MySQL no arquivo
                        filew.write(value)
                        filew.write("');\n")
            elif i == 3 or i == 4: # Se contador for 3 ou 4
                if value not in t2: # Se valor n√£o tiver sido gravado antes
                    t2.append(value) # Registra a grava√ß√£o do valor
                    if i == 3: # Se for 3 grava o valor com sintaxe MySQL no arquivo
                        idOrgaoSubordinado = value
                        filew.write("INSERT INTO orgaosubordinado (idOrgaoSubordinado, nomeOrgaoSubordinado, idOrgaoSuperior) ")
                        filew.write("VALUES (")
                        filew.write(value)
                        filew.write(", '")
                    elif i == 4: # Se for 4 grava o valor com sintaxe MySQL no arquivo
                        filew.write(value)
                        filew.write("', ")
                        filew.write(idOrgaoSuperior)
                        filew.write(");\n")
            elif i == 11 or i == 12:
                if value not in t3:
                    t3.append(value)
                    if i == 11:
                        idFuncao = value
                        filew.write("INSERT INTO funcao (idFuncao, nomeFuncao) VALUES (")
                        filew.write(value)
                        filew.write(", '")
                    elif i == 12:
                        filew.write(value)
                        filew.write("');\n")

                        filew.write("INSERT INTO orgaosubordinado_funcao (idOrgaoSubordinado, idFuncao) VALUES (")
                        filew.write(idOrgaoSubordinado)
                        filew.write(", ")
                        filew.write(idFuncao)
                        filew.write(");\n")
            elif i == 13 or i == 14:
                if value not in t4:
                    t4.append(value)
                    if i == 13:
                        filew.write("INSERT INTO subfuncao (idSubFuncao, nomeSubFuncao) VALUES (")
                        filew.write(value)
                        filew.write(", '")
                        idSubFuncao = value
                    elif i == 14:
                        filew.write(value)
                        filew.write("');\n")
                        
                        filew.write("INSERT INTO subfuncao_funcao (idSubFuncao, idFuncao) VALUES (")
                        filew.write(idSubFuncao)
                        filew.write(", ")
                        filew.write(idFuncao)
                        filew.write(");\n")
            elif i == 15 or 16:
                if value not in t5:
                    t5.append(value)
                    if i == 15:
                        filew.write("INSERT INTO programa (idPrograma, nomePrograma) VALUES (")
                        filew.write(value)
                        filew.write(", '")
                        idPrograma = value
                    elif i == 16:
                        filew.write(value)
                        filew.write("');\n")
                        
                        filew.write("INSERT INTO subfuncao_programa (idSubFuncao, idPrograma) VALUES (")
                        filew.write(idSubFuncao)
                        filew.write(", ")
                        filew.write(idPrograma)
                        filew.write(");\n")
            
            if i == 24 or i == 25:
                if i == 24:
                    dataPgto = substitoi(value) # Data de pagamento passa por uma fun√ß√£o para coloc√°-la no formato adequado
                if dataPgto != False: # Se o valor n√£o for uma data, retornar√° falso e o valor n√£o ser√° registrado
                    # H√° v√°rias inser√ß√µes no arquivo que suas datas foram classificadas como sigilosas e n√£o possuem o
                    # formato de data, e sim de texto. Por isso a cria√ß√£o do IF anterior
                    if i == 24: # Se i for 24
                        filew.write("INSERT INTO valorpago (datapagamento, valor, idValorPago) VALUES ('")
                        filew.write(substitoi(value))
                        filew.write("', ")
                    if i == 25: # Se i for 25, o valor a ser gravado deve ser o valor pago
                        done = value.strip() # Por ser o √∫ltimo da linha do CSV vem com um catact√©r de quebra de linha
                        # Esse o motivo da fun√ß√£o anterior
                        idV = str(idValorPago) # O id do valor pago deve ser uma string para a inser√ß√£o no arquivo final
                        filew.write(done.replace(',','.')) 
                        filew.write(", ")
                        filew.write(idV) # Grava o ID
                        filew.write(");\n")

                        filew.write("INSERT INTO programa_valorpago (idPrograma, idValorPago) VALUES(")
                        filew.write(idPrograma)
                        filew.write(", ")
                        filew.write(idV)
                        filew.write(");\n")

                        idValorPago += 1 # Incrementa o ID. No arquivo o ID n√£o √© fornecido
                                         # criamos aqui apenas para inser√ß√£o no banco.
            i+=1
        print(counter,'/',counted) # Imprime a quantidade atual que j√° foram analizadas e a quantidade total
    #tempw.write()
    
# Fun√ß√£o que substitui o valor de data comum DD/MM/AAAA por AAAA-MM-DD aceito pelo MySQL
# Tamb√©m detecta se o valor inserido n√£o √© uma data no formato acima.
def substitoi(valor:str):
    try: # Tente...
        data = valor.split('/') # Dividir a string por barra e armazenar
        cont = 1 # Contador inicia
        dia = '' # Cria 3 strings
        mes = ''
        ano = ''
        for i in data: # para cada i em data
            if cont == 1:
                dia = str(int(i)) # Armazene o dia. Aqui se o valor n√£o for um n√∫mero, gerar√° um erro
            elif cont == 2:
                mes = i # Armazene o m√™s
            elif cont == 3:
                ano = i # Armazene o ano
            cont+=1
        return ano + '-' + mes + '-'+ dia # Retorne a data no formato aceito
    except: # Se o erro acontecer...
        return False # Retorne falso.
    
```
